```python
new_all = data[data['Order ID'].duplicated(keep = False)]
new_all.head()
```

```python
new_all['Product_Bundle'] = new_all.groupby('Order ID')['Item'].transform(lambda x:','.join(x))

new_all.head()
```

```python

new_all = new_all[['Order ID', 'Product_Bundle']].drop_duplicates()
new_all.head()
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in new_all['Product_Bundle']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list,2)))
    
    
top = count.most_common(10)
top
```

```python
val = []
label = []

for i in top:
    label.append(i)
    val.append(i[1])
print(label)
print(val)
# y = np.array([35, 25, 25, 15])
# mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(val, labels = label)
# plt.legend()
```
